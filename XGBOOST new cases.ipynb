```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import time
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator as gen
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
col_list=["date","new_cases","positive_rate"]
dataset = pd.read_csv("new cases dataset.csv",usecols=col_list)
dataset.describe()
```

```python
dataset['date']=pd.to_datetime(dataset['date'], infer_datetime_format= True)
df= dataset.loc[:,dataset.columns !='date']
```

```python
scaler= MinMaxScaler()

dataScaled=scaler.fit_transform(df)
```

```python
features= dataScaled
target= dataScaled[:,0]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(features, target, test_size= 0.10, random_state=123, shuffle=False) 
```

```python
model = XGBRegressor(n_estimators=500, learning_rate=0.01)
model.fit(X_train,Y_train, eval_set=[(X_train, Y_train),(X_test, Y_test)], eval_metric= 'mae')
```

```python
predictions =model.predict(X_test)
```

```python
predictions
```

```python
predictions.reshape(-1,1)
```

```python
predvtest = pd.concat([pd.DataFrame(Y_test), pd.DataFrame(predictions)], axis=1)
```

```python
rev_trans= scaler.inverse_transform(predvtest)
```

```python
rev_trans
```

```python
final= pd.DataFrame(rev_trans)
```

```python
final.columns=['new_cases','predicted']
final
```

```python
from sklearn.metrics import mean_absolute_error
mae_new_cases = mean_absolute_error(final['predicted'],final['new_cases'])
print('Mean absolute error of new cases is:',mae_new_cases)
```
