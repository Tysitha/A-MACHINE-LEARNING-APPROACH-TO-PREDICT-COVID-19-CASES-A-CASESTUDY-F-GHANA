```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVR
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
dataset = pd.read_csv("new cases dataset.csv", index_col='date', parse_dates=True)
print('shape',dataset.shape)
dataset.describe()
```

```python
X = dataset[['positive_rate']]
Y = dataset[['new_cases']]
scaler= MinMaxScaler()
X=scaler.fit_transform(X)
Y = scaler.fit_transform(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.10, random_state=123, shuffle=False) 
```

```python
from sklearn.svm import SVR
linSVR = SVR(kernel='linear')
linSVR.fit(X_train, Y_train.ravel())

polySVR = SVR(kernel='poly', degree=2)
polySVR.fit(X_train, Y_train.ravel())

rbfSVR = SVR(kernel='rbf', C=1000.0, gamma=0.15)
rbfSVR.fit(X_train, Y_train.ravel())
```

```python
Y_pred = polySVR.predict(X_test)
```

```python
y_testvy_pred =pd.concat([pd.DataFrame(Y_test), pd.DataFrame(Y_pred)], axis=1)
```

```python
rev_trans= scaler.inverse_transform(y_testvy_pred)
```

```python
rev_trans
```

```python
final= pd.DataFrame(rev_trans)
```

```python
final.columns=['new_cases','predicted']
final
```

```python
from sklearn.metrics import mean_absolute_error
mae_new_cases = mean_absolute_error(final['predicted'],final['new_cases'])
print('Mean absolute error of new cases is:',mae_new_cases)
```
